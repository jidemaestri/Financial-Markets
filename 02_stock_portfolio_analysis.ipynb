```python
stocks = ["AAPL", "GOOG", "MSFT", "VALE", "MELI", "DIS", "KO", "MCD", "NVDA", "PAM", "GGAL", "BBAR"]
```

```python
for stock in stocks:
    info = yf.Ticker(stock).info
    print(info["shortName"], info["regularMarketPrice"], info["sector"], info["country"])
```

```python
import pandas as pd
stock_portfolio = pd.DataFrame()
```

```python
# # Definir una función que convierta los números a valores en millones
# def to_millions(x):
#     return "{:.2f}M".format(x / 1000000)
```

```python

for stock in stocks:
    financials = yf.Ticker(stock).info

    name = financials["shortName"]
    current_price = financials["regularMarketPrice"]
    roa = financials["returnOnAssets"]
    fcf_margin = financials["freeCashflow"]
    gross_margin = financials["grossMargins"]
    current_ratio = financials["currentRatio"]
    free_cash_flow = financials["freeCashflow"]
    pb_ratio = financials["priceToBook"]
    revenue_growth = financials["revenueGrowth"] 
    earningsGrowth = financials["earningsGrowth"]
    targetLowPrice = financials["targetLowPrice"]
    targetMedianPrice = financials["targetMedianPrice"]
    targetHighPrice = financials["targetHighPrice"]
    net_profit_margin = financials["netIncomeToCommon"] / financials["totalRevenue"]
    #evfcf_ratio = financials["enterpriseValue"]/financials["freeCashflow"]
    debt_to_equity = financials["debtToEquity"]
    priceToSalesTrailing12Months = financials["priceToSalesTrailing12Months"]
    twoHundredDayAverage  = financials["twoHundredDayAverage"]
    totalCashPerShare = financials["totalCashPerShare"]

    
    
    stock_portfolio = stock_portfolio.append({"Stock": name, 
                    "Current Price": current_price,
                    "Debt/Equity": debt_to_equity, 
                    "Current Ratio": current_ratio, 
                    "Price To Sales 12M Ratio": priceToSalesTrailing12Months, 
                    #"EV/FCF Ratio": evfcf_ratio, 
                    "P/B Ratio": pb_ratio, 
                    "Revenue Growth": revenue_growth, 
                    "ROA": roa, 
                    "target High Price": targetHighPrice,
                    "target Median Price": targetMedianPrice,
                    "target Low Price": targetLowPrice,
                    "Gross Margin": gross_margin, 
                    "Free Cash Flow": free_cash_flow,
                    "FCF Margin": fcf_margin, 
                    "Net Profit Margin": net_profit_margin,
                    "total Cash Per Share": totalCashPerShare, 
                    "twoHundred Day Avg": twoHundredDayAverage,
                    },
                   ignore_index=True)

    
    



    
```
